```python
df['Class'].value_counts()
```

```python
df_majority = df[(df['Class']==0)]
df_minority = df[(df['Class']==1)]
```

```python
from sklearn.utils import resample
```

```python
#upsample minority class
df_minority_upsample = resample(df_minority,n_samples=100000,replace=True)
df_upsampled = pd.concat([df_minority_upsample,df_majority])
```

```python
df_upsampled['Class'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x = df_upsampled.iloc[:,:-1]
```

```python
x
```

```python
y = df_upsampled.iloc[:,-1]
```

```python
y
```

```python
scalar = StandardScaler()
```

```python
x = scalar.fit_transform(x)
```

```python
X_tr,X_test,y_tr,y_test = train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_tr,y_tr)
```

```python
pred = rfc.predict(X_test)
```

```python
accuracy_score(y_test,pred)
```
